```python
def ToWeight(y):
    w = np.zeros(y.shape, dtype=float)
    ind = y != 0
    w[ind] = 1./(y[ind]**2)
    return w

def rmspe(yhat, y):
    w = ToWeight(y)
    rmspe = np.sqrt(np.mean( w * (y - yhat)**2 ))
    return rmspe

def rmspe_xg(yhat, y):
    y = y.get_label()
    y = np.exp(y) - 1
    yhat = np.exp(yhat) - 1
    w = ToWeight(y)
    rmspe = np.sqrt(np.mean(w * (y - yhat)**2))
    return "rmspe", rmspe
```

# Modelling

### XGBoost

```python
import xgboost as xgb
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
```

```python
# Prepare the data
data = df.loc[(df['is_train'] == 1) & (df['Open'] == 1) & (df['Outlier'] == False)]
x_train, x_test, y_train, y_test = train_test_split(data[features_x], 
                                                    data[features_y], 
                                                    test_size=0.1, 
                                                    random_state=42)

# Option 1: Impute Missing Values
imputer = SimpleImputer(strategy='mean')
x_train = imputer.fit_transform(x_train)
x_test = imputer.transform(x_test)

# Convert the data into DMatrix format, which XGBoost requires
dtrain = xgb.DMatrix(x_train, label=y_train.values.ravel())
dtest = xgb.DMatrix(x_test, label=y_test.values.ravel())

# Set up parameters and train the model
param = {
    'max_depth': 9,
    'eta': 0.01,
    'subsample': 0.75,
    'colsample_bytree': 0.6,
    'objective': 'reg:squarederror',
}

num_round = 20000
evallist = [(dtrain, 'train'), (dtest, 'test')]

model = xgb.train(param, dtrain, num_round, evallist, 
                  early_stopping_rounds=250, verbose_eval=250)

# Predict and evaluate
y_pred_xgb = model.predict(dtest)
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
print(f"XGBoost RMSE: {rmse_xgb}")

# Feature Importance Visualization
xgb.plot_importance(model)
plt.show()

# Prepare Submission Data
submit = df.loc[df['is_train'] == 0]

# Apply the same imputation to the submission data
submit[features_x] = imputer.transform(submit[features_x])

# Convert submission data to DMatrix
dsubmit = xgb.DMatrix(submit[features_x])

# Generate predictions
predictions = model.predict(dsubmit)

# Prepare the submission DataFrame
df_predictions = submit['Id'].reset_index()
df_predictions['Id'] = df_predictions['Id'].astype('int')
df_predictions['Sales'] = (np.exp(predictions) - 1) * 0.985  # Scale back

# Sort by Id and save to CSV
df_predictions.sort_values('Id', inplace=True)
df_predictions[['Id', 'Sales']].to_csv('solution.csv', index=False)

print("Submission file 'solution.csv' has been created.")
```

```python
# Prepare the data
data = df.loc[(df['is_train'] == 1) & (df['Open'] == 1) & (df['Outlier'] == False)]
x_train, x_test, y_train, y_test = train_test_split(data[features_x], 
                                                    data[features_y], 
                                                    test_size=0.1, 
                                                    random_state=42)

# Impute missing values
imputer = SimpleImputer(strategy='mean')
x_train = imputer.fit_transform(x_train)
x_test = imputer.transform(x_test)

# Initialize and train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, max_depth=9, random_state=42)
rf_model.fit(x_train, y_train.values.ravel())

# Predict and evaluate on the test set
y_pred_rf = rf_model.predict(x_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print(f"Random Forest RMSE: {rmse_rf}")

# Feature Importance Visualization
plt.figure(figsize=(18, 8))
plt.barh(data[features_x].columns, rf_model.feature_importances_)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Random Forest Feature Importance')
plt.show()
```
